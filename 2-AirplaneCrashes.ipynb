## üìå ADIM 3: VERƒ∞ TEMƒ∞ZLEME (Data Cleaning)

### üéØ Ama√ß
Eksik deƒüerleri, dubliko kayƒ±tlarƒ± ve hatalarƒ± d√ºzeltmek.

### üìö Teori
**Veri Temizleme ƒ∞≈ülemleri:**
1. **Dubliko Deƒüerler** - Aynƒ± kazanƒ±n 2 kere kaydedilmesi
2. **Eksik Deƒüerler** - NULL/NaN deƒüerleri
3. **Aykƒ±rƒ± Deƒüerler** - Mantƒ±k dƒ±≈üƒ± rakamlar
4. **Yanlƒ±≈ü Veri T√ºrleri** - Tarih yerine metin yazƒ±lmasƒ±

**Neden √ñnemli?**
- Yanlƒ±≈ü temizlenmi≈ü veri ‚Üí Yanlƒ±≈ü sonu√ßlar
- "Veri Bilimi %80 temizleme, %20 analiz" 

### üíª Uygulama

````python
print("\n" + "=" * 80)
print("ADIM 3: VERƒ∞ TEMƒ∞ZLEME")
print("=" * 80)

# Veri setinin orijinal boyutu
print(f"\nüìä Orijinal veri seti boyutu: {df.shape[0]} satƒ±r")

# 1Ô∏è‚É£ DUBLIKO KAYITLAR
print("\n1Ô∏è‚É£ DUBLIKO (KOPƒ∞) KAYITLAR")
print("-" * 80)

duplicate_count = df.duplicated().sum()
print(f"\n   Toplam dubliko satƒ±r: {duplicate_count}")

if duplicate_count > 0:
    df = df.drop_duplicates()
    print(f"   ‚úÖ Dubliko satƒ±rlar silindi!")
    print(f"   üîÑ Yeni veri seti boyutu: {df.shape[0]} satƒ±r")
else:
    print(f"   ‚úÖ Y√ºksek sevin√ß! Dubliko kayƒ±t yok!")

# 2Ô∏è‚É£ S√úTUN ƒ∞Sƒ∞MLERƒ∞Nƒ∞ STANDARTLA≈ûTIRMA
print("\n2Ô∏è‚É£ S√úTUN ƒ∞Sƒ∞MLERƒ∞Nƒ∞ TEMIZLE")
print("-" * 80)

print(f"\n   Eski s√ºtun isimleri:")
for i, col in enumerate(df.columns[:3], 1):
    print(f"      {i}. '{col}'")

# S√ºtun isimlerini temizle (bo≈üluklarƒ± kaldƒ±r, standart hale getir)
df.columns = df.columns.str.strip()

print(f"\n   ‚úÖ S√ºtun isimleri standardize edildi!")

# 3Ô∏è‚É£ EKSƒ∞K DEƒûER STRATEJƒ∞Sƒ∞
print("\n3Ô∏è‚É£ EKSƒ∞K DEƒûERLERI KALDIRMA STRATEJƒ∞Sƒ∞")
print("-" * 80)

print(f"\n   Strateji: N√úMERƒ∞K s√ºtunlardaki eksik deƒüerleri kaldƒ±r")

# Eksik deƒüer sayƒ±sƒ± √∂ncesi
rows_before = len(df)

# N√ºmerik s√ºtunlarda eksik deƒüerleri kaldƒ±r
numeric_cols = df.select_dtypes(include=[np.number]).columns
df = df.dropna(subset=numeric_cols, how='any')

rows_after = len(df)
rows_removed = rows_before - rows_after

print(f"\n   Silinen satƒ±r sayƒ±sƒ±: {rows_removed}")
print(f"   ‚úÖ Yeni veri seti boyutu: {rows_after} satƒ±r")

# 4Ô∏è‚É£ AYKIRI DEƒûERLER (OUTLIERS)
print("\n4Ô∏è‚É£ AYKIRI DEƒûERLER (OUTLIERS) TESPƒ∞Tƒ∞")
print("-" * 80)

print(f"\n   IQR (Interquartile Range) y√∂ntemi kullanacaƒüƒ±z")

for col in ['Fatalities', 'Aboard']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    print(f"\n   üìä {col}")
    print(f"      Normal aralƒ±k: {lower_bound:.0f} - {upper_bound:.0f}")
    print(f"      Aykƒ±rƒ± deƒüer sayƒ±sƒ±: {len(outliers)}")
    print(f"      √ñrnek aykƒ±rƒ± deƒüerler: {outliers[col].head(3).values}")

# 5Ô∏è‚É£ SONU√á
print("\n" + "=" * 80)
print("‚úÖ VERƒ∞ TEMƒ∞ZLEME SONU√áLARI")
print("=" * 80)
print(f"\n   Ba≈ülangƒ±√ß: {rows_before:,} kaza kaydƒ±")
print(f"   Sonu: {rows_after:,} kaza kaydƒ±")
print(f"   Temizlenen: {rows_removed} satƒ±r")
print(f"   Kalite: Veri temizlendi ve analiz i√ßin hazƒ±r!")

print(f"\nüí° Sonraki adƒ±mda: ƒ∞statistiksel analizler yapacaƒüƒ±z")
````

## üìå ADIM 4: ƒ∞STATƒ∞STƒ∞KSEL ANALƒ∞Z (Statistical Analysis)

### üéØ Ama√ß
Veri hakkƒ±nda derinlemesine matematiksel analiz yaparak √∂l√ß√ºlebilir bulgular elde etmek.

### üìö Teori
**Temel ƒ∞statistikler:**
- **Merkezi Eƒüilim**: Ortalama, medyan, mod
- **Yayƒ±lƒ±m**: Standart sapma, varyans, aralƒ±k
- **Daƒüƒ±lƒ±m ≈ûekli**: √áarpƒ±klƒ±k (skewness), basƒ±klƒ±k (kurtosis)
- **Frekans Analizi**: Hangi deƒüerler en sƒ±k olu≈üuyor?

**√ñrnek:**
- "Ortalama √∂l√ºm sayƒ±sƒ± 80 ama √ßoƒüu kaza 10-20 ki≈üi √∂l√ºm" ‚Üí Saƒüa √ßarpƒ±k daƒüƒ±lƒ±m
- Aykƒ±rƒ± b√ºy√ºk kazalar daƒüƒ±lƒ±mƒ± skew yapƒ±yor

### üíª Uygulama

````python
print("\n" + "=" * 80)
print("ADIM 4: ƒ∞STATƒ∞STƒ∞KSEL ANALƒ∞Z")
print("=" * 80)

# 1Ô∏è‚É£ MERKEZI Eƒûƒ∞Lƒ∞M √ñL√á√úLERƒ∞
print("\n1Ô∏è‚É£ MERKEZI Eƒûƒ∞Lƒ∞M √ñL√á√úLERƒ∞ (Central Tendency)")
print("-" * 80)
print("\n   Bu √∂l√ß√ºler verilerin hangi noktada toplandƒ±ƒüƒ±nƒ± g√∂sterir:")

numeric_cols_clean = ['Fatalities', 'Aboard']

for col in numeric_cols_clean:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode()[0] if len(df[col].mode()) > 0 else None
    
    print(f"\n   üìä {col}")
    print(f"      ‚Ä¢ Ortalama (Mean): {mean_val:,.1f}")
    print(f"      ‚Ä¢ Medyan (Median): {median_val:,.1f}")
    print(f"      ‚Ä¢ Mod (Most Common): {mode_val:,.0f}")
    
    if mean_val > median_val:
        print(f"      üí° Mean > Median ‚Üí SAƒûA √áARPIK daƒüƒ±lƒ±m (aykƒ±rƒ± y√ºksek deƒüerler var)")

# 2Ô∏è‚É£ YAYILIM √ñL√á√úLERƒ∞
print("\n" + "=" * 80)
print("2Ô∏è‚É£ YAYILIM √ñL√á√úLERƒ∞ (Dispersion)")
print("-" * 80)
print("\n   Bu √∂l√ß√ºler verilerin ne kadar yayƒ±lƒ± olduƒüunu g√∂sterir:")

for col in numeric_cols_clean:
    std_val = df[col].std()
    var_val = df[col].var()
    range_val = df[col].max() - df[col].min()
    
    print(f"\n   üìä {col}")
    print(f"      ‚Ä¢ Standart Sapma (Std Dev): {std_val:,.1f}")
    print(f"      ‚Ä¢ Varyans: {var_val:,.1f}")
    print(f"      ‚Ä¢ Aralƒ±k (Range): {range_val:,.0f} (Min: {df[col].min():.0f} - Max: {df[col].max():.0f})")
    
    cv = (std_val / df[col].mean()) * 100
    print(f"      ‚Ä¢ Deƒüi≈ükenlik Katsayƒ±sƒ±: {cv:.1f}%")

# 3Ô∏è‚É£ DAƒûILIM ≈ûEKLƒ∞
print("\n" + "=" * 80)
print("3Ô∏è‚É£ DAƒûILIM ≈ûEKLƒ∞ (Distribution Shape)")
print("-" * 80)
print("\n   √áarpƒ±klƒ±k ve Basƒ±klƒ±k - Daƒüƒ±lƒ±m ne kadar \"normal\" g√∂r√ºn√ºyor?")

for col in numeric_cols_clean:
    skewness = stats.skew(df[col].dropna())
    kurtosis = stats.kurtosis(df[col].dropna())
    
    print(f"\n   üìä {col}")
    print(f"      ‚Ä¢ √áarpƒ±klƒ±k (Skewness): {skewness:.3f}")
    
    if skewness > 1:
        print(f"         ‚ûú SAƒûA √áARPIK (right-skewed): Saƒüda uzun kuyruk")
    elif skewness < -1:
        print(f"         ‚ûú SOLA √áARPIK (left-skewed): Solda uzun kuyruk")
    else:
        print(f"         ‚ûú NISPETEN SIMETRIK")
    
    print(f"      ‚Ä¢ Basƒ±klƒ±k (Kurtosis): {kurtosis:.3f}")
    if kurtosis > 3:
        print(f"         ‚ûú SIVRI daƒüƒ±lƒ±m (aykƒ±rƒ± deƒüerler daha fazla)")
    else:
        print(f"         ‚ûú D√úZGEN daƒüƒ±lƒ±m")

# 4Ô∏è‚É£ FREKANS ANALƒ∞Zƒ∞
print("\n" + "=" * 80)
print("4Ô∏è‚É£ FREKANS ANALƒ∞Zƒ∞ (Frequency Analysis)")
print("-" * 80)
print("\n   Hangi deƒüerler en sƒ±k tekrarlanƒ±yor?")

if 'Date' in df.columns:
    print(f"\n   üìÖ En √ßok kaza olan TARƒ∞HLER:")
    try:
        # Tarihleri parse et ve yƒ±l √ßƒ±kar
        df['Year'] = pd.to_datetime(df['Date'], errors='coerce').dt.year
        top_years = df['Year'].value_counts().head(5)
        for year, count in top_years.items():
            print(f"      ‚Ä¢ {int(year)}: {count} kaza")
    except:
        print(f"      ‚Ä¢ (Tarih formatƒ± ayƒ±klanamadƒ±)")

# 5Ô∏è‚É£ √ñZET ƒ∞STATƒ∞STƒ∞KLER
print("\n" + "=" * 80)
print("5Ô∏è‚É£ DETAYLI ƒ∞STATƒ∞STƒ∞KSEL √ñZET")
print("=" * 80)
print(df[numeric_cols_clean].describe().to_string())
````

### üéØ Ama√ß

### üìö Teori
**Neden G√∂rselle≈ütirme?**
- ƒ∞nsan beyni resimler 1000 kelimeden daha hƒ±zlƒ± i≈üler
- Sayƒ±lardan kalƒ±plar g√∂rmek zor, grafikten kolay
- Sunumlar ve raporlar i√ßin gerekli

**Grafik T√ºrleri:**
- **Histogram**: Daƒüƒ±lƒ±mƒ± g√∂rmek
- **Scatter Plot**: ƒ∞li≈ükileri g√∂rmek

## üìå ADIM 6: VERƒ∞ HAZIRLAMASI (Preprocessing)

### üéØ Ama√ß
Makine √∂ƒürenmesi modelleri i√ßin verileri uygun formata d√∂n√º≈üt√ºrmek.

### üìö Teori
**Neden Preprocessing?**
- Modeller sayƒ±sal veriler ister (metin deƒüil)
- Modeller farklƒ± √∂l√ßekleri karƒ±≈ütƒ±rƒ±r (1-100 vs 1-10000)
- Train/Test ayrƒ±mƒ± = model deƒüerlendirmesinin saƒülƒ±ƒüƒ±

**StandardScaler Nedir?**
- Verileri "normalize" eder (ortalama=0, std=1)
- KNN, SVM gibi mesafe temelli modellere gerekli
- T√ºm √∂zellikleri e≈üit aƒüƒ±rlƒ±kta yapar

### üíª Uygulama

````python
print("\n" + "=" * 80)
print("ADIM 6: VERƒ∞ HAZIRLAMASI (Preprocessing)")
print("=" * 80)

# 1Ô∏è‚É£ FEATURES VE TARGET'I AYIR
print("\n1Ô∏è‚É£ √ñZELLƒ∞KLERƒ∞ (FEATURES) VE HEDEFƒ∞ (TARGET) AYIR")
print("-" * 80)

# NaN deƒüerleri kaldƒ±r
df_clean = df.dropna(subset=['Aboard', 'Fatalities']).copy()

# Hed–µf deƒüi≈ükeni olu≈ütur: √ñl√ºm sayƒ±sƒ±na g√∂re kategorize et
# Hafif: 0-10, Orta: 11-50, Aƒüƒ±r: 51+
df_clean['Severity'] = pd.cut(df_clean['Fatalities'], bins=[0, 10, 50, float('inf')], 
                        labels=['Hafif', 'Orta', 'Aƒüƒ±r'])

# NaN deƒüerleri sil (pd.cut'dan kaynaklanan)
df_clean = df_clean.dropna(subset=['Severity'])

# √ñzellikler (Features)
X = df_clean[['Aboard']].values
# Hedef (Target)
y = df_clean['Severity'].values

print(f"\n   üìä √ñzellikler (X):")
print(f"      ≈ûekil: {X.shape}")

print(f"\n   üéØ Hedef (y):")
print(f"      ≈ûekli: {y.shape}")
print(f"      Sƒ±nƒ±flar: {np.unique(y)}")
print(f"\n      Sƒ±nƒ±f Daƒüƒ±lƒ±mƒ±:")
unique, counts = np.unique(y, return_counts=True)
for severity, count in zip(unique, counts):
    percent = (count / len(y)) * 100
    print(f"      ‚Ä¢ {severity}: {count:,} kaza ({percent:.1f}%)")

# 2Ô∏è‚É£ √ñL√áEKLENDIRME
print("\n" + "=" * 80)
print("2Ô∏è‚É£ VERƒ∞LERƒ∞ √ñL√áEKLENDIR (StandardScaler)")
print("-" * 80)

print(f"\n   √ñl√ßeklendirme √∂ncesi:")
print(f"      Min: {X[:, 0].min():.0f} | Max: {X[:, 0].max():.0f}")
print(f"      Ortalama: {X[:, 0].mean():.0f} | Std: {X[:, 0].std():.0f}")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"\n   √ñl√ßeklendirme sonrasƒ±:")
print(f"      Min: {X_scaled[:, 0].min():.2f} | Max: {X_scaled[:, 0].max():.2f}")
print(f"      Ortalama: {X_scaled[:, 0].mean():.2f} | Std: {X_scaled[:, 0].std():.2f}")
print(f"\n   ‚úÖ Veriler standart normal daƒüƒ±lƒ±ma √∂l√ßeklendirildi!")

# 3Ô∏è‚É£ EƒûITIM-TEST B√ñL√úM√ú
print("\n" + "=" * 80)
print("3Ô∏è‚É£ VERƒ∞LERƒ∞ EƒûITIM VE TEST SETINE B√ñL (%80-%20)")
print("-" * 80)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\n   Toplam veri: {len(X_scaled):,}")
print(f"\n   üè´ Eƒüitim Seti (%80):")
print(f"      Satƒ±r sayƒ±sƒ±: {len(X_train):,}")

print(f"\n   ‚úÖ Test Seti (%20):")
print(f"      Satƒ±r sayƒ±sƒ±: {len(X_test):,}")

print(f"\n   üí° Stratified Split: Her sƒ±nƒ±f e≈üit oranda train/test'e ayrƒ±ldƒ±")

print(f"\n‚úÖ ADIM 6 TAMAMLANDI: Veriler modelleme i√ßin hazƒ±r!")
````

## üìå ADIM 7: MAKINE √ñƒûRENMESƒ∞ (Machine Learning)

### üéØ Ama√ß
Eƒüitim verilerinden √∂r√ºnt√ºler √∂ƒürenerek yeni verilere tahmin yapan modeller olu≈üturmak.

### üìö Teori
**Model Nedir?**
- Eƒüitim verilerindeki matematiksel ili≈ükileri √∂ƒürenen sistem
- Yeni veri g√∂rd√ºƒü√ºnde tahmin yapabiliyor
- √ñrnek: "U√ßakta 100 ki≈üi varsa kaza aƒüƒ±r olur" kuralƒ±nƒ± √∂ƒüreniyor

**Hangi Modeller?**
1. **K-Nearest Neighbors (KNN)**: En yakƒ±n kom≈üulara bakƒ±p oy ister
2. **Random Forest**: √áok sayƒ±da karar aƒüacƒ±nƒ± birle≈ütirir

**Deƒüerlendirme:**
- **Accuracy**: Ka√ß % doƒüru tahmin yaptƒ±?
- **Confusion Matrix**: Hangi sƒ±nƒ±flarƒ± karƒ±≈ütƒ±rƒ±yor?
- **Classification Report**: Detaylƒ± performans analizi

### üíª Uygulama

````python
print("\n" + "=" * 80)
print("ADIM 7: MAKINE √ñƒûRENMESƒ∞ (Classification)")
print("=" * 80)

# 1Ô∏è‚É£ K-NEAREST NEIGHBORS (KNN) MODELƒ∞
print("\n1Ô∏è‚É£ K-NEAREST NEIGHBORS (KNN) MODELƒ∞")
print("-" * 80)

print("\n   ü§ñ Model Tanƒ±mƒ±: 'Benzer kom≈üularƒ± bul ve onlarƒ±n sƒ±nƒ±fƒ±nƒ± al'")

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

knn_pred = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_pred)

print(f"\n   üìä KNN Modeli Performansƒ±:")
print(f"      ‚úÖ Doƒüruluk (Accuracy): {knn_accuracy:.4f} ({knn_accuracy*100:.2f}%)")
print(f"      ‚ùå Hata Oranƒ±: {1-knn_accuracy:.4f} ({(1-knn_accuracy)*100:.2f}%)")

print(f"\n   A√ßƒ±klama:")
print(f"      ‚Ä¢ Test setindeki {len(y_test)} kazanƒ±n {int(knn_accuracy*len(y_test))} tanesini doƒüru tahmin etti")

# 2Ô∏è‚É£ RANDOM FOREST MODELƒ∞
print("\n" + "=" * 80)
print("2Ô∏è‚É£ RANDOM FOREST MODELƒ∞")
print("-" * 80)

print("\n   ü§ñ Model Tanƒ±mƒ±: '√áok sayƒ±da karar aƒüacƒ±nƒ± kullan ve √ßoƒüunluk oyu al'")

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"\n   üìä Random Forest Modeli Performansƒ±:")
print(f"      ‚úÖ Doƒüruluk (Accuracy): {rf_accuracy:.4f} ({rf_accuracy*100:.2f}%)")
print(f"      ‚ùå Hata Oranƒ±: {1-rf_accuracy:.4f} ({(1-rf_accuracy)*100:.2f}%)")

print(f"\n   A√ßƒ±klama:")
print(f"      ‚Ä¢ {int(rf_accuracy*len(y_test))} doƒüru, {int((1-rf_accuracy)*len(y_test))} yanlƒ±≈ü tahmin")

# 3Ô∏è‚É£ MODEL KAR≈ûILA≈ûTIRMASI
print("\n" + "=" * 80)
print("3Ô∏è‚É£ MODELLERƒ∞ KAR≈ûILA≈ûTIR")
print("-" * 80)

models_results = {
    'K-Nearest Neighbors': knn_accuracy,
    'Random Forest': rf_accuracy
}

print(f"\n   Sƒ±ralanmƒ±≈ü Sonu√ßlar:")
for i, (model_name, accuracy) in enumerate(sorted(models_results.items(), 
                                                    key=lambda x: x[1], reverse=True), 1):
    stars = "‚≠ê" * (round(accuracy * 5))
    print(f"   {i}. {model_name:25s} ‚Üí {accuracy*100:6.2f}% {stars}")

best_model_name = max(models_results, key=models_results.get)
print(f"\n   üèÜ En ƒ∞yi Model: {best_model_name}")

# 4Ô∏è‚É£ DETAYLI ANALƒ∞Z
print("\n" + "=" * 80)
print("4Ô∏è‚É£ EN ƒ∞Yƒ∞ MODEL'ƒ∞N DETAYLI ANALƒ∞Zƒ∞")
print("-" * 80)

print(f"\n   Confusion Matrix ({best_model_name}):")
cm = confusion_matrix(y_test, rf_pred)
print(f"\n   {cm}")

print(f"\n   Classification Report:")
print(f"\n{classification_report(y_test, rf_pred)}")

# 5Ô∏è‚É£ √ñRNEK TAHMINLER
print("\n" + "=" * 80)
print("5Ô∏è‚É£ √ñRNEK TAHMƒ∞NLER")
print("-" * 80)

test_samples = [
    {"Aboard": 50, "explanation": "K√º√ß√ºk u√ßak"},
    {"Aboard": 200, "explanation": "B√ºy√ºk u√ßak"},
    {"Aboard": 300, "explanation": "√áok b√ºy√ºk u√ßak"}
]

print(f"\n   Model yeni veriler √ºzerinde tahmin yapƒ±yor:")

for sample in test_samples:
    sample_scaled = scaler.transform([[sample["Aboard"]]])
    
    knn_pred_sample = knn.predict(sample_scaled)[0]
    knn_proba = knn.predict_proba(sample_scaled)[0].max() * 100
    
    rf_pred_sample = rf.predict(sample_scaled)[0]
    rf_proba = rf.predict_proba(sample_scaled)[0].max() * 100
    
    print(f"\n   üìå Senaryo: {sample['explanation']} ({sample['Aboard']} ki≈üi)")
    print(f"      KNN Tahmini: {knn_pred_sample} (G√ºven: {knn_proba:.1f}%)")
    print(f"      RF Tahmini:  {rf_pred_sample} (G√ºven: {rf_proba:.1f}%)")

print(f"\n‚úÖ ADIM 7 TAMAMLANDI: Modeller ba≈üarƒ±yla eƒüitildi ve deƒüerlendirildi!")
````
