## üìå ADIM 6: VERƒ∞ HAZIRLAMASI (PREPROCESSING)

Makine √∂ƒürenmesi modellerini eƒüitmeden √∂nce verileri standartla≈ütƒ±rmamƒ±z gerekir.
Bu i≈ülem ≈üu nedenlerden √∂nemlidir:

1. **√ñl√ßeklendirme**: Farklƒ± √∂l√ßeklerdeki √∂zellikleri aynƒ± √∂l√ßeƒüe getirmek
2. **Normalizasyon**: Deƒüerleri belirli bir aralƒ±ƒüa (√∂rneƒüin 0-1) getirmek
3. **Eƒüitim-Test Ayrƒ±mƒ±**: Modeli deƒüerlendirmek i√ßin ayrƒ± test seti

### Neden √ñl√ßeklendirme √ñnemli?

Bazƒ± algoritmalar (KNN, SVM, Sinir Aƒülarƒ±) √∂zelliklerin √∂l√ßeƒüine duyarlƒ±dƒ±r.
- Sepal length: 4-8 cm aralƒ±ƒüƒ±nda
- Petal length: 1-7 cm aralƒ±ƒüƒ±nda

√ñl√ßeklendirme yapmazsak, petal length modeli daha fazla etkileyebilir!

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print("=" * 60)
print("VERƒ∞ HAZIRLAMASI (PREPROCESSING)")
print("=" * 60)

# 1Ô∏è‚É£ √ñzellikleri ve hedef deƒüi≈ükeni ayƒ±r
X = df.drop(['target', 'flower_type'], axis=1)  # √ñzellikler
y = df['target']  # Hedef deƒüi≈üken

print("\n1. √ñzellikleri ve Hedef Deƒüi≈ükeni Ayƒ±rma:")
print(f"   X ≈üekli (Features): {X.shape}")
print(f"   y ≈üekli (Target): {y.shape}")

# 2Ô∏è‚É£ Verileri standardize et
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\n2. √ñl√ßeklendirme Sonu√ßlarƒ±:")
print(f"   Orijinal X - Min: {X.values.min():.2f}, Max: {X.values.max():.2f}")
print(f"   √ñl√ßeklendirilmi≈ü X - Min: {X_scaled.min():.2f}, Max: {X_scaled.max():.2f}")
print("   ‚úÖ Veriler standart normal daƒüƒ±lƒ±ma √∂l√ßeklendirildi!")

# 3Ô∏è‚É£ Eƒüitim ve test setine b√∂l
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print(f"\n3. Eƒüitim-Test Ayrƒ±mƒ±:")
print(f"   Eƒüitim seti boyutu: {X_train.shape[0]} (%80)")
print(f"   Test seti boyutu: {X_test.shape[0]} (%20)")
```

---

## üìå ADIM 7: MAKƒ∞NE √ñƒûRENMESƒ∞ MODELLERƒ∞

Makine √∂ƒürenmesi modelleri, verilerden √∂r√ºnt√ºler √∂ƒürenen ve tahminler yapan algoritmalar olarak d√º≈ü√ºn√ºlebilir.

### 3 Pop√ºler Sƒ±nƒ±flandƒ±rma Algoritmasƒ±:

1. **K-Nearest Neighbors (KNN)**
   - Basit ve sezgisel
   - "Yakƒ±ndaki kom≈üulara bakar ve kendi grubuna katƒ±lƒ±r"
   - Hƒ±zlƒ± ama b√ºy√ºk veri setlerinde yava≈ü

2. **Logistic Regression**
   - Doƒürusal model
   - Olasƒ±lƒ±k tabanlƒ± tahminler
   - Hƒ±zlƒ± ve yorumlanabilir

3. **Random Forest**
   - Ensemble y√∂ntemi
   - Bir√ßok karar aƒüacƒ±nƒ± birle≈ütirir
   - Genellikle y√ºksek doƒüruluk

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

print("=" * 60)
print("MODELLER'ƒ∞ Eƒûƒ∞T VE DEƒûERLENDIR")
print("=" * 60)

# Model 1: K-Nearest Neighbors
print("\nü§ñ MODEL 1: K-NEAREST NEIGHBORS (KNN)")
print("-" * 60)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_pred)

print(f"Doƒüruluk (Accuracy): {knn_accuracy:.4f} ({knn_accuracy*100:.2f}%)")
print(f"Hata: {1-knn_accuracy:.4f} ({(1-knn_accuracy)*100:.2f}%)")

# Model 2: Logistic Regression
print("\n\nü§ñ MODEL 2: LOGƒ∞STƒ∞C REGRESSION")
print("-" * 60)

lr = LogisticRegression(max_iter=200, random_state=42)
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_pred)

print(f"Doƒüruluk (Accuracy): {lr_accuracy:.4f} ({lr_accuracy*100:.2f}%)")
print(f"Hata: {1-lr_accuracy:.4f} ({(1-lr_accuracy)*100:.2f}%)")

# Model 3: Random Forest
print("\n\nü§ñ MODEL 3: RANDOM FOREST")
print("-" * 60)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"Doƒüruluk (Accuracy): {rf_accuracy:.4f} ({rf_accuracy*100:.2f}%)")
print(f"Hata: {1-rf_accuracy:.4f} ({(1-rf_accuracy)*100:.2f}%)")
```

```python
# Modelleri Kar≈üƒ±la≈ütƒ±r
print("\n\n" + "=" * 60)
print("MODEL KAR≈ûILA≈ûTIRMASI")
print("=" * 60)

models_comparison = {
    'KNN': knn_accuracy,
    'Logistic Regression': lr_accuracy,
    'Random Forest': rf_accuracy
}

# Sonu√ßlarƒ± sƒ±ralƒ± ≈üekilde g√∂ster
for i, (model_name, accuracy) in enumerate(sorted(models_comparison.items(), 
                                                    key=lambda x: x[1], reverse=True), 1):
    print(f"{i}. {model_name:25s} - Doƒüruluk: {accuracy:.4f} ({accuracy*100:.2f}%)")

best_model = max(models_comparison, key=models_comparison.get)
print(f"\n‚úÖ En iyi model: {best_model} (Doƒüruluk: {models_comparison[best_model]:.2%})")

# En iyi modelin detaylƒ± analizi (Random Forest)
print("\n\n" + "=" * 60)
print("EN ƒ∞Yƒ∞ MODEL: RANDOM FOREST - DETAYLI ANALIZ")
print("=" * 60)

print("\nKonf√ºzyon Matrisi (Confusion Matrix):")
cm = confusion_matrix(y_test, rf_pred)
print(cm)
print("\nSatƒ±rlar: Ger√ßek deƒüer (Actual)")
print("S√ºtunlar: Tahmin (Predicted)")

print("\n\nSƒ±nƒ±flandƒ±rma Raporu (Classification Report):")
print(classification_report(y_test, rf_pred, target_names=['Setosa', 'Versicolor', 'Virginica']))
```

```python
# √ñzellik √ñnemliliƒüi (Feature Importance)
print("\n" + "=" * 60)
print("√ñZELLƒ∞K √ñNEMLƒ∞Lƒ∞ƒûƒ∞ (FEATURE IMPORTANCE)")
print("=" * 60)

feature_importance = pd.DataFrame({
    '√ñzellik': X.columns,
    '√ñnem': rf.feature_importances_
}).sort_values('√ñnem', ascending=False)

print("\nRandom Forest modeli hangi √∂zellikleri daha √∂nemli buluyor?")
print(feature_importance.to_string(index=False))

# G√∂rselle≈ütir
plt.figure(figsize=(10, 6))
bars = plt.barh(feature_importance['√ñzellik'], feature_importance['√ñnem'], color='steelblue')
plt.xlabel('√ñnem Skoru')
plt.ylabel('√ñzellikler')
plt.title('Random Forest - √ñzellik √ñnemliliƒüi')
plt.xlim(0, feature_importance['√ñnem'].max() * 1.1)

# Deƒüerleri barlarƒ±n √ºzerine yaz
for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, 
             f'{width:.3f}', ha='left', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("\nüìå A√ßƒ±klama: Petal √∂zellikleri, √ßi√ßek t√ºr√ºn√º belirlemede daha √∂nemlidir!")
```

## üìå ADIM 8: PRATIK √ñRNEK - YENƒ∞ VERƒ∞ TAHMƒ∞N ETME

Eƒüittiƒüimiz model, hi√ß g√∂rmediƒüi yeni √ßi√ßek √∂rneklerinin t√ºr√ºn√º tahmin edebilir.

```python
print("=" * 60)
print("YENƒ∞ √áƒ∞√áEK T√úR√úN√ú TAHMIN ET")
print("=" * 60)

# √ñrnek 1: Setosa benzeri √∂l√ß√ºmler
new_flower_1 = [[5.1, 3.5, 1.4, 0.2]]  # Sepal Length, Sepal Width, Petal Length, Petal Width
new_flower_1_scaled = scaler.transform(new_flower_1)
prediction_1 = rf.predict(new_flower_1_scaled)[0]
prediction_1_proba = rf.predict_proba(new_flower_1_scaled)[0]

print("\nüå∏ √ái√ßek 1 √ñl√ß√ºmleri:")
print(f"   Sepal Length: {new_flower_1[0][0]} cm")
print(f"   Sepal Width: {new_flower_1[0][1]} cm")
print(f"   Petal Length: {new_flower_1[0][2]} cm")
print(f"   Petal Width: {new_flower_1[0][3]} cm")
print(f"\n   Tahmin: {['Setosa', 'Versicolor', 'Virginica'][prediction_1]}")
print(f"   G√ºvenilirlik:")
for i, (flower, prob) in enumerate(zip(['Setosa', 'Versicolor', 'Virginica'], prediction_1_proba)):
    print(f"      {flower:12s}: {prob:.2%}")

# √ñrnek 2: Virginica benzeri √∂l√ß√ºmler
new_flower_2 = [[7.2, 3.2, 6.0, 1.8]]
new_flower_2_scaled = scaler.transform(new_flower_2)
prediction_2 = rf.predict(new_flower_2_scaled)[0]
prediction_2_proba = rf.predict_proba(new_flower_2_scaled)[0]

print("\n\nüå∏ √ái√ßek 2 √ñl√ß√ºmleri:")
print(f"   Sepal Length: {new_flower_2[0][0]} cm")
print(f"   Sepal Width: {new_flower_2[0][1]} cm")
print(f"   Petal Length: {new_flower_2[0][2]} cm")
print(f"   Petal Width: {new_flower_2[0][3]} cm")
print(f"\n   Tahmin: {['Setosa', 'Versicolor', 'Virginica'][prediction_2]}")
print(f"   G√ºvenilirlik:")
for i, (flower, prob) in enumerate(zip(['Setosa', 'Versicolor', 'Virginica'], prediction_2_proba)):
    print(f"      {flower:12s}: {prob:.2%}")
```

### √ñnemli Konseptler:

| **Standardization** | Verileri ortalama=0, std=1 olacak ≈üekilde √∂l√ßeklendirme |

### Iris Veri Seti Bulgularƒ±:
