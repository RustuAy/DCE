## DA-5: 3) ASSESS — Veri kalitesi & hazırlık

Açıklama (kısa): Eksik değerleri (missing) ve veri türlerini kontrol ederim. Hızlı stratejiler: sayısal eksikleri median ile doldurmak; kategoriklerde 'Unknown' veya bir ek flag kullanmak.

Mülakat notu: "Eksik verinin yapısını inceleyip (örn. hangi sütunlar çok eksik) uygun strateji uygularım; örneğin `Age` için median doldurma, `Cabin` çok eksikse onu feature engineering ile kullanabilirim (ör: 'Cabin_present')."

### Mülakat Soruları — ASSESS
- **Soru:** "Eksik verilerle nasıl uğraşırsınız?"
  - **Cevap:** "İlk önce eksiklerin dağılımını incelerim (hangi sütunlar, ne kadar). Sayısal eksikler için median/mean, kategorik için en sık kategori ya da 'Unknown' seçeneğini kullanırım; çok eksik sütunlar için drop veya farklı engineering düşünürüm."

- **Soru:** "Neden median kullanıyorsun, mean yerine?"
  - **Cevap:** "Median uç değerlere daha dayanıklıdır; çarpık dağılımlarda median daha güvenilir sonuç verir."

````python
# ASSESS: Hızlı kalite kontrolleri ve basit düzeltmeler
# Açıklama: df.info() ve eksik değerler, ardından Age ve Fare için doldurma örneği, Cabin için presence flag

print('Özet info:')
print(df.info())

print('\nEksik değer sayıları:')
print(df.isnull().sum())

# Age median ile doldurma
age_median = df['Age'].median()
df['Age_filled'] = df['Age'].fillna(age_median)

# Fare eksik bir değer var - median ile dolduralım
fare_median = df['Fare'].median()
df['Fare_filled'] = df['Fare'].fillna(fare_median)

# Cabin çok eksik, bunun yerine 'Cabin_present' flag'i ekleyelim
df['Cabin_present'] = df['Cabin'].notnull().astype(int)
# Eğer Cabin'leri kategorik olarak kullanmak istersek (eksik->'Unknown')
df['Cabin_filled'] = df['Cabin'].fillna('Unknown')

print('\nAfter imputation:')
print(df[['Age','Age_filled','Fare','Fare_filled','Cabin_present']].head())
print('\nEksik sayilari (son hali):')
print(df[['Age_filled','Fare_filled','Cabin_filled']].isnull().sum())

# Basit doğrulama: Age_filled'de eksik olmadığını kontrol et
assert not df['Age_filled'].isnull().any(), 'Age_filled hala eksik var!'
print('\nAge_filled eksik değil, devam edebiliriz.')
````

## DA-5: 4) ANALYZE — Keşifsel Veri Analizi (EDA) ve Basit Model

Açıklama: Sınıf ve cinsiyete göre hayatta kalma oranlarını inceleyip basit görselleştirme yapacağız. Ardından Logistic Regression pipeline ile basit bir sınıflandırma modeli kurup doğruluğu hesaplayacağız.

Mülakat notu: "Önce verilere bakarım (Pclass, Sex gibi feature'lar genelde güçlü gösterge olur), sonra baseline model kurarım; performans yetersizse feature engineering üzerinde çalışırım."

  - **Cevap:** "Başlangıçta basit, yorumlanabilir modeller (logistic regression, decision tree) ile başlarım; baseline kurar, sonra gerekirse daha karmaşık modellere geçerim."

````python
# ANALYZE: EDA görselleştirmeleri ve pivotlar
import matplotlib.pyplot as plt
import seaborn as sns

# Pclass'e göre hayatta kalma oranı
pclass_surv = df.groupby('Pclass')[TARGET].mean()
print('Survival rate by Pclass:\n', pclass_surv)

plt.figure(figsize=(6,4))
sns.barplot(x=pclass_surv.index, y=pclass_surv.values)
plt.title('Survival rate by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival rate')
plt.show()

# Sex'e göre hayatta kalma
sex_surv = df.groupby('Sex')[TARGET].mean()
print('\nSurvival rate by Sex:\n', sex_surv)

plt.figure(figsize=(6,4))
sns.barplot(x=sex_surv.index, y=sex_surv.values)
plt.title('Survival rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival rate')
plt.show()
````

````python
# ANALYZE: Basit model pipeline (impute + one-hot + logistic regression)
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

features = ['Pclass','Sex','Age_filled','Fare_filled','Cabin_present']
X = df[features]
y = df[TARGET]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

numeric_feats = ['Age_filled','Fare_filled']
cat_feats = ['Pclass','Sex']
pre = ColumnTransformer([
    ('num', SimpleImputer(strategy='median'), numeric_feats),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_feats)
])

pipe = make_pipeline(pre, LogisticRegression(max_iter=1000))
pipe.fit(X_train, y_train)

pred = pipe.predict(X_test)
acc = accuracy_score(y_test, pred)
print('Test accuracy:', round(acc,3))
print('\nClassification report:\n', classification_report(y_test, pred, zero_division=0))

# Mülakat notu: model doğruluğunu baseline ile karşılaştır (ör: most frequent)
from sklearn.dummy import DummyClassifier
dummy = DummyClassifier(strategy='most_frequent')
dummy.fit(X_train, y_train)
print('Baseline accuracy (most frequent):', round(dummy.score(X_test,y_test),3))

# Eğer acc > KPI_ACCURACY, hedefe ulaşmış sayılabilir
print('KPI hedefi:', KPI_ACCURACY)
````

````python
# METRİK ÖRNEĞİ: Confusion matrix ve temel metriklerin hesaplanması
# Açıklama: Aşağıdaki kod confusion matrix, accuracy, precision, recall, f1 ve AUC'yi hesaplar
# ve her birinin kısa Türkçe açıklamasını ekranda gösterir.
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, accuracy_score

# Test set üzerinde tahminlerimiz zaten 'pred' ile hesaplanmıştı
cm = confusion_matrix(y_test, pred)
acc = accuracy_score(y_test, pred)
prec = precision_score(y_test, pred, zero_division=0)
rec = recall_score(y_test, pred, zero_division=0)
f1 = f1_score(y_test, pred, zero_division=0)
# AUC için olasılık tahminleri gerektiğinden proba yoksa ROC AUC hesaplaması atlanabilir
try:
    proba = pipe.predict_proba(X_test)[:,1]
    auc = roc_auc_score(y_test, proba)
except Exception:
    auc = None

print('Confusion Matrix:\n', cm)
print('\nAccuracy (Doğruluk):', round(acc,3), "— Tüm tahminler içinde doğru olanların oranı.")
print('Precision (Kesinlik):', round(prec,3), "— Pozitif tahminlerin kaçının doğru olduğu (TP/(TP+FP)).")
print('Recall (Duyarlılık):', round(rec,3), "— Gerçek pozitiflerin kaçının yakalandığı (TP/(TP+FN)).")
print('F1 Score:', round(f1,3), "— Precision ve Recall'un harmonik ortalaması.")
if auc is not None:
    print('AUC (ROC altındaki alan):', round(auc,3), "— Modelin sınıfları ayırt etme kabiliyeti (1.0 en iyi).")
else:
    print('AUC hesaplanamadı (proba gerekli).')

# Not: zero_division=0 kullanıldı; pratikte bu durum veri dağılımına göre yorumlanmalı.
````

### 'Sex' özelliğini çıkarıp modeli yeniden eğitme ve metrikleri karşılaştırma

Aşağıdaki hücre `Sex` sütununu çıkarır, modeli yeniden eğitir ve test metriklerini orijinal modelle karşılaştırır. Bu, `Sex` özelliğinin model performansına etkisini göstermek için öğretici amaçlıdır.

````python
# 'Sex' olmadan model (Türkçe)
# Açıklama: 'Sex' özelliğini çıkarıp modeli yeniden eğitiyoruz; metrikleri karşılaştırıp kısa yorum veriyoruz.
# Önceki metrikleri al (önceki hücrede hesaplanmıştı)
prev_metrics = dict(acc=acc, prec=prec, rec=rec, f1=f1, auc=auc)

# 'Sex' olmadan kullanılacak özellikler
features_no_sex = ['Pclass','Age_filled','Fare_filled','Cabin_present']
X_ns = df[features_no_sex]
from sklearn.model_selection import train_test_split, cross_val_score

# Eğitim/test split
Xn_train, Xn_test, yn_train, yn_test = train_test_split(X_ns, y, random_state=42, stratify=y)

numeric_feats = ['Age_filled','Fare_filled']
cat_feats_ns = ['Pclass']
pre_ns = ColumnTransformer([
    ('num', SimpleImputer(strategy='median'), numeric_feats),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_feats_ns)
])

pipe_ns = make_pipeline(pre_ns, LogisticRegression(max_iter=1000))
pipe_ns.fit(Xn_train, yn_train)

# Tahmin ve metrikler
pred_ns = pipe_ns.predict(Xn_test)
acc_ns = accuracy_score(yn_test, pred_ns)
prec_ns = precision_score(yn_test, pred_ns, zero_division=0)
rec_ns = recall_score(yn_test, pred_ns, zero_division=0)
f1_ns = f1_score(yn_test, pred_ns, zero_division=0)
try:
    proba_ns = pipe_ns.predict_proba(Xn_test)[:,1]
    auc_ns = roc_auc_score(yn_test, proba_ns)
except Exception:
    auc_ns = None

print('Sex olmadan model - Test doğruluğu:', round(acc_ns,3))
print('Precision (Kesinlik):', round(prec_ns,3), 'Recall (Duyarlılık):', round(rec_ns,3), 'F1:', round(f1_ns,3))
if auc_ns is not None:
    print('AUC:', round(auc_ns,3))
else:
    print('AUC hesaplanamadı (olasılık tahmini gerekebilir)')

print('\nOrijinal model (Sex ile):', prev_metrics)
print('\nSex olmadan model için 5-fold CV doğrulukları:')
cv_ns = cross_val_score(pipe_ns, X_ns, y, cv=5, scoring='accuracy')
print([round(s,3) for s in cv_ns], 'ortalama:', round(cv_ns.mean(),3))

# Katsayılar (feature etkileri)
try:
    feat_names_ns = pre_ns.get_feature_names_out()
    coef_ns = pipe_ns.named_steps['logisticregression'].coef_[0]
    feat_coef_ns = list(zip(feat_names_ns, coef_ns))
    feat_coef_ns_sorted = sorted(feat_coef_ns, key=lambda x: -abs(x[1]))
    print('\nÖnemli özellikler (isim, katsayı) — Sex yok:')
    for n,c in feat_coef_ns_sorted:
        print(f" - {n}: {round(c,3)}")
except Exception as e:
    print('Feature isimleri çıkarılamadı:', e)

# Kısa Türkçe yorum
if prev_metrics['acc'] == 1.0 and acc_ns < 1.0:
    print('\nNot: Orijinal modelde görülen mükemmel sonuçların birincil nedeni `Sex` özelliğinin güçlü ayırıcı olmasıdır; `Sex` çıkarıldığında performans düştü, bu da özelliğin etkisini gösterir.')
else:
    print('\nNot: Karşılaştırma yaparken orijinal modelin ve yeni modelin performansını dikkatle yorumlayın; farklı feature kombinasyonları deneyin.')
````

## Ölçütler (Metrikler) — Kısa Türkçe Açıklamalar
- **Accuracy (Doğruluk):** Tüm tahminler içinde doğru olanların oranı. Kolay anlaşılır ama sınıflar dengesizse yanıltıcı olabilir.
- **Precision (Kesinlik):** Modelin pozitif tahminlerinin ne kadarının doğru olduğunu gösterir. (TP / (TP + FP))
- **Recall (Duyarlılık):** Gerçek pozitiflerin ne kadarının doğru tahmin edildiğini gösterir. (TP / (TP + FN))
- **F1 Skoru:** Precision ve Recall'un harmonik ortalamasıdır; dengesiz sınıflarda tek bir özet gösterge sunar.
- **AUC (Area Under ROC Curve):** ROC eğrisi altındaki alan; modelin pozitif ve negatif sınıfları ayırt etme kabiliyetini ölçer (1.0 en iyi, 0.5 rastgele tahmin seviyesinde).
- **Baseline (Referans):** Modelin performansını değerlendirmek için kullanılan basit bir metot (örn. en sık sınıfı tahmin etme).
